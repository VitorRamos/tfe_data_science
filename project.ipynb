```python
import pandas as pd
import inspect,os,matplotlib
import matplotlib.pyplot as plt
matplotlib.use('TkAgg')
from tkinter import tix
from pylab import plot,axis,savefig,show,title
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg, NavigationToolbar2Tk)
from matplotlib.figure import Figure
from matplotlib.widgets import Slider
root=tix.Tk()
def main():
    file='interface_setup.txt'
    this_file=inspect.getfile(inspect.currentframe())
    this_folder=os.path.dirname(os.path.abspath(this_file))
    setup=setupFile()
    graphe=graph()
    setup.read(this_folder+os.sep+file)
    setup._colorframe='#4f81bd'
    root.configure(bg=setup._colorframe)
    mainwindow(root,setup,graphe)
    root.mainloop()
class graph():
    def __init__(self):
        self.fig=Figure(figsize=(12,8))
        self.ax1 = self.fig.add_subplot(111)
    def sliders_on_changed(val):
        line.set_ydata(val)
        self.fig.canvas.draw_idle()
    def simpleGraphe(self,master):
        canvas=FigureCanvasTkAgg(self.fig,master=master)
        canvas.get_tk_widget().place(x=300,y=40)
        canvas.draw()
        a0 = 5
        f0 = 3
        delta_f = 5.0
        axcolor = 'lightgoldenrodyellow'
        ax=self.fig.add_axes([0.25, 0.15, 0.65, 0.03])
        sfreq=Slider(ax,'Freq',0.1, 30.0, valinit=f0)
        sfreq.on_changed(self.sliders_on_changed)
class setupFile():
    def __init__(self):
        self._colorbutton='#ffffff'
        self._colorflow='#ffffff'
        self._colorframe='#ffffff'
    def read(self,filename):
        GoodFile=False
        filetype=''
        comment=False
        tempbool=False
        with open(filename,'r')as f:
            for line_idx,line in enumerate(f):
                temp=line.replace('>'," ")
                tempo=temp.replace('<',' ')
                y=tempo.lower()
                x=y.split(' ')
                if ('filetype'in x)or (GoodFile==True):
                    if 'filetype'in x:
                        filetype=x[6]
                        if filetype=='interface_setup':
                            GoodFile=True
                    if 'colorframe'in x:
                        self._colorframe=x[6]
                    if 'colorbutton'in x:
                        self._colorbutton=x[6]
                    if 'colorflow'in x:
                        self._colorflow=x[6]
                elif (filetype!='')and(filetype!='interface_setup'):
                    return('interface_setup')
                    break

class mainwindow():
    def __init__(self,master,setup,graphe):
        self.setup=setup
        self.graphe=graphe
        self.master=master
        self.initGUI()
    def initGUI(self):
        RWidth=self.master.winfo_screenwidth()
        RHeight=self.master.winfo_screenheight()
        self.master.geometry(str(RWidth)+"x"+str(RHeight))
        self.master.title('Test')
        self.master.configure(bg=self.setup._colorframe)
        self.graphes()
    def graphes(self):
        self.graphe.simpleGraphe(self.master)
if __name__=='__main__':
    main()
```
